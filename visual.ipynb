```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [100, 150, 200, 250, 300],
    'Profit': [20, 30, 50, 70, 90]
})

# Plotting both Sales and Profit over time
data.plot(x='Year', y=['Sales', 'Profit'], kind='line', title="Sales and Profit Over Time")
```

```python
data = pd.DataFrame({
    'usage_of_dictionary': [0, 50, 100, 20, 30, 0],
    'figure_it_on_your_own': [100, 70, 40, 90, 80, 100],
    'materials': ['python language', 'data structures', 'chatgpt', 'mongodb', 'psql', 'gql'],
})
data.plot(
    x="materials",
    y=["usage_of_dictionary", "figure_it_on_your_own"], 
    title="tomer's plot",
    kind="line",
    rot=45
)
    
```

```python
data = pd.DataFrame({
    'lecturers': ['enosh', 'tomer', 'matanel', 'itshak', 'shmuel'],
    'muscles': [100.0, 20.2, 47.90, 01.1, 68.0],
    'afraid_of_shmuel': [0.0, 22.2, 8.8, 89.90, 60.0],
})
data.plot(
    x="lecturers",
    y=["muscles", "afraid_of_shmuel"], 
    title="who's afraid of the shmuel",
    kind="line"
)
```

```python
import matplotlib.pyplot as plt

data = pd.read_json('./students.json')
```

```python
data.plot(
    x="Name",
    y=["Class_Average", "Peer_Support_Count"],
    title="Class Average vs Peer Support Count by Student",
    kind="line",
    rot=45
)
plt.tight_layout()
plt.show()
```

```python
data.plot(
    x="Age",
    y="Class_Average",
    title="Age vs Class Average",
    kind="scatter"
)
plt.tight_layout()
plt.show()
```

```python
data.plot(
    x="Name",
    y=["Late_Submissions", "Peer_Support_Count"],
    title="Late Submissions vs Peer Support Count",
    kind="bar"
)
plt.tight_layout()
plt.show()
```

```python
data.plot(
    x="Name",
    y="Age",
    title="Age Distribution of Students",
    kind="bar"
)
plt.tight_layout()
plt.show()
```

```python
data.sort_values('Class_Average', ascending=False).plot(
    x="Name",
    y="Class_Average",
    title="Class Average by Student (Sorted)",
    kind="bar"
)
plt.tight_layout()
plt.show()
```
